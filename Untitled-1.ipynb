### scaler data

```python
#lstm (la fonction sigmoïde prend des valeurs entre 0 et 1)
# il serait préférable de mettre à l'échelle nos valeurs entre 0 et 1, il est donc préférable d'utiliser MinMaxScaler 
#pas standarscaler
df_lstm=data.copy()
size=int(len(df_lstm)*0.8)
train =df_lstm.iloc[0:size] 
test = df_lstm.iloc[size:] 
```

```python

train.shape,test.shape
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train = scaler.fit_transform(train)
scaled_test = scaler.transform(test)
y_train = scaler.fit_transform(train["Open"].values.reshape(-1, 1)).reshape(-1, )
y_test = scaler.transform(test["Open"].values.reshape(-1, 1)).reshape(-1, )
 
```

### LSTM

```python
n_input = 30
batch_size=1
n_features=train.shape[1]
#TimeseriesGenerator for train data
train_generator = TimeseriesGenerator(scaled_train ,y_train, length=n_input, batch_size=batch_size)
#TimeseriesGenerator for test data
test_generator = TimeseriesGenerator(scaled_test,y_test, length=n_input,batch_size=batch_size)
```

```python
def model_lstm(input1,input2,input3):
  model=Sequential() 
  model.add(LSTM(input1,activation='relu',input_shape=(n_input,n_features),return_sequences=True, name="lstm_1"))
  #model.add(Dropout(0.2))
  model.add(LSTM(input2,activation='relu',return_sequences=True, name="lstm_2"))
  #model.add(Dropout(0.2))
  model.add(LSTM(input3,activation='relu',return_sequences=False, name="lstm_3")) 
 # model.add(Dropout(0.2))
#  model.add(LSTM(input4,activation='relu',return_sequences=False, name="lstm_4"))
  model.add(Dense(1, name="dense"))
  model.compile(optimizer="adam", loss='mse', metrics=['mae'])
  return model
 
model=model_lstm(128,64,32)
model.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                                    patience=2,
                                                    mode='min')
```

```python
history = model.fit(train_generator, epochs=20,
                    validation_data=test_generator,
                    shuffle=False)
                    #callbacks=[early_stopping])
```

```python
model.evaluate(test_generator, verbose=0) 
```

```python
model.evaluate(train_generator, verbose=0) 
```

```python
# Creating trace1
trace1 = go.Scatter(
                    x= history.epoch,
                    y = history.history["loss"],
                     
                    mode = "lines",
                    name = "train loss",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    
                    )
# Creating trace1
trace2 = go.Scatter(
                    x= history.epoch,
                    y = history.history["val_loss"],
  
                    mode = "lines+markers",
                    name = "test loss",
                    marker = dict(color='rgba(12, 50, 196, 0.6)'),
                  
                    )
data = [trace1, trace2]
  
layout = dict(title = 'Loss function mse',
              xaxis= dict(title= 'epoch',ticklen= 5,zeroline= False)
              
             )
  
fig = dict(data = data, layout = layout)
iplot(fig)
```

```python
def get_y_from_generator(gen):
    '''
    Get all targets y from a TimeseriesGenerator instance.
    '''
    y = None
    for i in range(len(gen)):
        batch_y = gen[i][1]
        if y is None:
            y = batch_y
        else:
            y = np.append(y, batch_y)
    y = y.reshape((-1,1))
    print(y.shape)
    return y
```

```python
trainY = get_y_from_generator(train_generator)
testY = get_y_from_generator(test_generator)
```

```python
trainPredict = model.predict(train_generator)
testPredict = model.predict(test_generator)
```
